```python
df["TotalPrice"] = df["UnitPrice"] * df["Quantity"]
df.head(5)
```

```python
total_invoices = df.groupby("InvoiceNo").sum()
total_invoices
```

```python
print ("số lượng hóa đơn bán ra: "+ str (total_invoices.size))
print ("Tổng doanh thu: " + str(total_invoices["TotalPrice"].sum()))
```

```python
quantity_product = total_invoices["Quantity"].sort_values(ascending = False)
quantity_product.head(10)
```
