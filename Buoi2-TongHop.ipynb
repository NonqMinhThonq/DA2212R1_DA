```python
#Tạo Dataframe từ list
lst = ['Kế toán','Kinh doanh','Tiếp thị']
df1 = pd.DataFrame(lst)
df1
```

```python
#Tạo dataframe từ list có nhiều hơn 2 cột và 2 dòng
#1 phần tử trong list sẽ là giá trị của một dòng(hoặc 1 điểm dữ liệu)
#Đối với điểm dữ liệu bị thiếu giá trị, giá trị mặc định sẽ là NaN
lst = [("Kế Toán", 20, 5), ("Giam Doc", 50), ("Nhan Vien",10)]
df1 = pd.DataFrame(lst, columns = ["Nghề","Lương","Phép"])
df1
```

```python
#Tạo Dataframe từ tuple
#pd.DataFrame đối với tuple chúng ta sẽ define columns ở tham số thứ 2 để tạo tên cho cột trong dataframe
#Nếu như chúng ta để trống columns, tên của cột tự động generate theo index
tuples = (('Kế toán', 2000),( 'Kinh doanh', 2100), ('Tiếp thị', 1900))
df2 = pd.DataFrame(tuples, columns = ["Nghề", "Lương"])
df2 
```

```python
#Tạo dataframe từ dictionary 
#Đối với dictionary, dataframe sẽ nhận key là cột, và value là dòng
#Đối với dictionary, giá trị value phải được để trong [] dù chỉ với 1 giá trị
dic = {"Nghề":["GD", "NhanVien", "Ke Toan"]
       ,"Lương":[50, 20,30]}
df3 = pd.DataFrame(dic)
df3
```

```python
#Tạo Series từ dictionary 
#Series là một cột của dataframe, chỉ có một chiều dữ liệu
#Tạo series từ dictionary, mỗi cặp key:value tương ứng với giá trị(một điểm dữ liệu) trong series
dic = {"Nghề":["GD", "NhanVien", "Ke Toan"]
       ,"Lương":[50, 20,30]}
df4 = pd.Series(dic)
df4
```

```python
#Tạo series từ array
#Series không cho phép chúng ta tạo từ dữ liệu có trên 2 chiều 
arr_error = np.array([[60,70,80,90],[6,7,8,9]])
arr = np.array([6,7,8,9])
df5 = pd.Series(arr)
df5
```

```python
#Tạo series từ các giá trị vô hướng
#Vậy tham số thứ nhất là giá trị truyền vào của Series
#Vậy tham số thứ hai là index mong muốn và số lượng điểm dữ liệu mà series có
series = pd.Series(10, index = [0,2,4])
series
```

```python
#Xuất dataframe ra các định dạng khác nhau
#Xuất dataframe ra CSV, tham số đầu tiên là tên file
df.to_csv('AmesHousing-CSV.csv')

#Xuất dataframe ra Excel,tham số đầu tiên là tên file
df.to_excel('AmesHousing-EXCEL.xlsx')

#Xuất dataframe ra JSon,tham số đầu tiên là tên file
#Đối với json, tham số thứ 2 sẽ là orient = "indexes" nếu là series và orient = "columns" nếu là dataframe
df.to_json('AmesHousing-JSON.json', orient = "columns")
           
#Xuất dataframe ra hdf5,tham số đầu tiên là tên file, tham số thứ 2 là format = "table"
df.to_hdf('AmesHousing-H5.h5', key = "str", format = 'table')

#Xuất dataframe ra HTML, tham số đầu tiên là tên file
df.to_html('AmesHousing-HTML.html')
```
